```python
%matplotlib inline
import numpy as np
from skimage.io import imread, imshow
img = imread('tiger-border.png')
imshow(img)
```

```python
img.shape
```

```python
borderColor = img[0, 0]
right = 0 
left = 0 
top = 0 
bottom = 0

for i in reversed(range(img.shape[0])):
    for j in reversed(range(img.shape[1])):
        if not np.array_equal(img[i,j], borderColor):
            top = i
            left = j
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if not np.array_equal(img[i,j], borderColor):
            bottom = img.shape[0] - i - 1
            right = img.shape[1] - j - 1
```

```python
print(left, top, right, bottom)
```
