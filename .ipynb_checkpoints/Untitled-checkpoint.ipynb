```python
import numpy as np
from skimage.io import imread, imsave,imshow 
from skimage.util import img_as_float, img_as_ubyte
img = imread('img3-2.png')
```

```python
imshow(img)
```

```python
R = img_as_float(img[:,:,0])
G = img_as_float(img[:,:,1])
B = img_as_float(img[:,:,2])
```

```python
y = 0.2126 * R + 0.7152 * G + 0.0722 * B
u = -0.0999 * R - 0.3360 * G + 0.4360 * B
v = 0.6150 * R - 0.5586 * G - 0.0563 * B
```

```python
x_min = np.percentile(y.reshape(-1), 5)
x_max = np.percentile(y.reshape(-1), 95)
y = (y - x_min) * (1 /(x_max - x_min))
y = np.clip(y, 0, 1.0)
```

```python
y_to_r = y + 1.2803 * v
u_to_g = y - 0.2148 * u - 0.3805 * v
v_to_b = y + 2.1279 * u
```

```python
R = img_as_ubyte(np.clip(y_to_r, 0, 1.0))
G = img_as_ubyte(np.clip(u_to_g, 0, 1.0))
B = img_as_ubyte(np.clip(v_to_b, 0, 1.0))
```

```python
out_img = np.dstack((R,G,B))
```

```python
imshow(out_img)
```
