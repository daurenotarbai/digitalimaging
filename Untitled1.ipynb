```python
from skimage.io import imread, imsave, imshow
from skimage import img_as_float, img_as_ubyte
from numpy import dstack, clip, mean
img = img_as_float(imread('img-3-3.png'))
imshow(img)
```

```python
R_avg = mean(img[:,:,0])
G_avg = mean(img[:,:,1])
B_avg = mean(img[:,:,2])
```

```python
avg = (R_avg + G_avg + B_avg) / 3
```

```python
rw = R_avg / avg
gw = G_avg / avg
bw = B_avg / avg
```

```python
R = img[:,:,0] / rw
G = img[:,:,1] / gw
B = img[:,:,2] / bw
```

```python
img_combined = dstack((R, G, B))
out_image = img_as_ubyte(clip(img_combined, 0, 1.0))
imshow(out_image)
```
