```python
from skimage.io import imread, imshow, imsave
img = imread('img.png')
imshow(img)
```

```python
img.shape[1]
```

```python
widthHalf = int(img.shape[0] / 2)
heightHalf = int(img.shape[1] / 2)
img[widthHalf, heightHalf] = [3,3,3]
imsave('out_img.png', img)
```
