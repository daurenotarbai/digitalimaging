```python
from PIL import Image, ImageFilter, ImageDraw, ImageFont
from skimage.io import imread, imshow, imsave
```

```python
original = Image.open("tiger-color.png")  
print("Размер изображения: ")  
print(original.format, original.size)
original.show()
```

```python
img = Image.open("tiger-color.png")
cropp=img.crop((200,300,400,500))
cropp.save("crop.png")
imshow("crop.png")
```

```python
img = Image.open("tiger-color.png")
idraw=ImageDraw.Draw(img)
headline=ImageFont.truetype('arial.ttf', size=30)
text='Tiger'
idraw.text((400,100),text,font=headline,fill = "green")
idraw.ellipse((300,220,350,250),fill = "green")
idraw.ellipse((470,220,520,250),fill = "green")
img.save("new_img.png")
imshow("new_img.png")
```
